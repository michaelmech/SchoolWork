```python
import pandas as pd
import numpy as np
##convert the already modified google sheet to a csv, then load it(the sheet doesn't have interrupted responses) 
Modified=pd.read_csv('Sorted_Tracksheet-Non_Colored.csv')
```

### Cleaning the Data to have Consistently-Typed Ethnicities

```python
ethnic_list=list(Modified['Ethnicity']) #list of all submitted ethnicities
Uethnicities=np.unique(ethnic_list) #array of unique submitted ethnicities

#possible cases of ethnicities transformed into the desired ethnicities
east_asian=['Asian','Chinese','Korean','Taiwanese','korean','East asian','asian','Cambodian','Filipino', 'Southeast Asian', 'Cambodian','filipino','Vietnamese',
            'Indonesian','East Asian ','South-East Asian','korean', 'Asian ', ' Asian',' Asian '];
south_asian=['Indian','Bangladeshi','Indian ','Asian Indian']
white=['Russian','white','Turkish']
hispanic=['Latino','Mexican','HIspanic','Latina','Hispanic ','Mexican American','hispanic']
black=['African American','North African']
middle_eastern=['MiddleEastern', 'middle eastern','Persian'] 
other= ['Asian and White','Asian-White', 'Asian/Mixed','Black & White', 'Black and Hispanic', 'Black and White',
       'Brazilian/White', 'East Asian and White',
       'East Asian/White', 'HIspanic and Asian', 'Hispanic/East Asian', 'Hispanic/Latino','Indian/South Asian', 'Latino/White',
        'Mexican and Filipina', 'Mixed/Other','Other','Other( South East asian','White & Filipino', 'White and Asian', 'White and East Asian',
       'White x East Asian', 'White&Black', 'White/East Asian', 'black and latina',
       'White/Hispanic']
```

```python
#modify to have all ethnicities be consistently typed

corrected=Modified.replace(to_replace=south_asian,value='South Asian')
corrected=corrected.replace(to_replace=east_asian,value='East Asian')
corrected=corrected.replace(to_replace=middle_eastern,value='Middle Eastern')
corrected=corrected.replace(to_replace=black,value= 'Black')
corrected=corrected.replace(to_replace=white,value='White')
corrected=corrected.replace(to_replace=hispanic,value='Hispanic')
corrected=corrected.replace(to_replace=other,value='Other')
```
